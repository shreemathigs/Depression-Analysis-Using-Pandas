## Analysis of the Depression Data

**Depression Rate in %**

```python
(data['Depression'].mean() * 100).round(2)
```

**How many Males and Females are depressed?**

```python
data.groupby(['Gender'])['Depression'].sum()
```

**Depression count based on City**

```python
data.groupby(['City'])['Depression'].sum().sort_values(ascending = False).head(10)
```

**Depression count based on family history of mental illness**

```python
data.groupby(['Mental illness'])['Depression'].count()
```

**Depression count based on suicidal thoughts**

```python
data.groupby(['Suicidal thoughts'])['Depression'].count()
```

**Sleep Duration based on Profession count**

```python
data.groupby(['Sleep Duration'])['Profession'].count()
```

**Sleep Duration based on average Age**

```python
data[data['Depression']==1].groupby(['Sleep Duration'])['Age'].mean().round(0).astype(int)
```

**Professions who are affected by Depression based on average age**

```python
data[data['Depression']==1 ].groupby(['Profession']) [['Age']].mean().round(0).astype(int)
```

**Sleep Duration of the Profession based on average age**

```python
data.groupby(['Sleep Duration','Profession'])[['Age']].mean().round().astype(int)
```

```python
# Creating of new column Age Group
def Age_group(x):
  if x <= 18:
    return 'Teen'
  if x <= 25:
    return '18-25'
  if x <=35:
    return '26-35'
  if x <= 45:
    return '36-45'
  else:
    return '50+'
data['Age_group']=data['Age'].apply(Age_group)
data.head()
```

## Analysing of the Highest depression rate

**Which age facing of the highest and lowest depression rate?**

```python
(data.groupby(['Age'])['Depression'].mean()*100).round(2)
```

**Which age group has the highest depression rate?**

```python
(data.groupby(['Age_group']) ['Depression'].mean()*100).round(2).sort_values(ascending = False)
```

**Which city has the highest number of depressed people?**

```python
# Top 5 cities of depressed people
data[data['Depression']==1].groupby(['City'])['Depression'].count().sort_values(ascending=False).head(5)
```

**Which Profession has the highest number of depression?**

```python
# Top 5 Profession has depression
data[data['Depression']==1].groupby(['Profession'])['Depression'].count().sort_values(ascending = False).head(5)
```

**Which degree has the highest number of depression?**

```python
data[data['Depression']==1].groupby(['Degree'])['Depression'].count().sort_values(ascending = False).head(10)
```

**CGPA based on Depression**

```python
data['CGPA_Category']= data['CGPA'].apply(lambda x: 'Low' if x<6 else 'High')
(data.groupby(['CGPA_Category'])['Depression'].mean()*100).round(2)
```

**What is the average Academic pressure and CGPA of students who are having 9+ hours of study hours?**

```python
data[
    (data['Profession']=='Student') &
     (data['Work/Study Hours'] >= 9)] [['Academic Pressure','CGPA']].mean().round(2)
```

**What are the  levels of job satisfaction and financial stress for professional with work pressure?**

```python
data[ (data['Profession']!='Students') &
     (data['Work Pressure'] > 3)] [['Job Satisfaction','Financial Stress']]
```

**How many people feeling suicidal thoughts, mental illness due to depression?**

```python
(data.groupby(['Suicidal thoughts','Mental illness'])['Depression'].mean()*100).sort_values(ascending = False).round(2)
```

**Profession who follows of healthy dietary habits**

```python
(data[data['Dietary Habits']=='Healthy'].groupby(['Profession']) ['Depression'].mean()*100).sort_values(ascending = False).round(2)
```
